```python
# Display image

import cv2
img = cv2.imread('Sample.jpg')
desired_width = 800
desired_height = 600

# Step 3: Resize the image
resized_image = cv2.resize(img, (desired_width, desired_height))
cv2.imshow('Image', resized_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Convert the image to grayscale
gray_img = cv2.cvtColor(resized_image, cv2.COLOR_BGR2GRAY)

cv2.imshow('Image', gray_img)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Convert the image to RGB
rgb_img = cv2.cvtColor(resized_image, cv2.COLOR_BGR2RGB)
cv2.imshow('Grayscale Image', rgb_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# Smoothing and Sharpening of Image

```python
blurred_img = cv2.GaussianBlur(resized_image, (9, 9), 0)
cv2.imshow('Blurred Image', blurred_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import numpy as np


sharpening_kernel = np.array([[-1, -1, -1],
                              [-1, 9, -1],
                              [-1, -1, -1]])

sharpened_img = cv2.filter2D(resized_image, -1, sharpening_kernel)
```

```python
cv2.imshow('Sharpered Image', sharpened_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Inserting Noise
import numpy as np


image = resized_image


gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)


mean = 0
variance = 0.1
sigma = variance ** 0.5
gaussian_noise = np.random.normal(mean, sigma, gray_image.shape)
gaussian_noise = gaussian_noise.reshape(gray_image.shape).astype(np.uint8)


noisy_image = cv2.add(gray_image, gaussian_noise)


cv2.imshow('Original Image', gray_image)
cv2.imshow('Noisy Image', noisy_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python

noisy_image = resized_image


gray_image = cv2.cvtColor(noisy_image, cv2.COLOR_BGR2GRAY)


blurred_image = cv2.GaussianBlur(gray_image, (5, 5), 0)


cv2.imshow('Noisy Image', gray_image)
cv2.imshow('Denoised Image', blurred_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np

# Load the image
image = resized_image


b, g, r = cv2.split(image)

# Create an array of zeros with the same shape as the image channels
zeros = np.zeros_like(b)

# Create windows to display the RGB channels
cv2.namedWindow('Red Channel', cv2.WINDOW_NORMAL)
cv2.namedWindow('Green Channel', cv2.WINDOW_NORMAL)
cv2.namedWindow('Blue Channel', cv2.WINDOW_NORMAL)

# Display each channel frame by frame
for frame in [r, g, b]:
    cv2.imshow('Red Channel', cv2.merge([frame, zeros, zeros]))
    cv2.imshow('Green Channel', cv2.merge([zeros, frame, zeros]))
    cv2.imshow('Blue Channel', cv2.merge([zeros, zeros, frame]))
    cv2.waitKey(8000)  # Wait for 1 second
    cv2.destroyAllWindows()

# Close the windows after displaying all frames
cv2.destroyAllWindows()
```
